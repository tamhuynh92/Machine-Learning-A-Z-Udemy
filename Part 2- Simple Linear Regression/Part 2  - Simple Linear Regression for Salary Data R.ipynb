```r
#Read the dataset
dataset = read.csv('Salary_Data.csv')

#Splitting the dataset into the training set and test set
library(caTools)
set.seed(123)
split = sample.split(dataset$Salary, SplitRatio = 2/3)
training_set = subset(dataset, split == TRUE)
test_set = subset(dataset, split == FALSE)
```

```r
# Fitting Simple Linear Regression to Training test
regressor = lm(formula = Salary ~ YearsExperience, data = training_set)
```

```r
y_pred = predict(regressor, newdata = test_set)
y_pred
```

```r
#Visualizing the training set results
library(ggplot2)

pl1 = ggplot() + geom_point(aes(x = training_set$YearsExperience, y = training_set$Salary), col = 'red') 
pl2 = pl1 + geom_line(aes(x = training_set$YearsExperience, y = predict(regressor, newdata = training_set)), col = 'blue')
pl3 = pl2 + ggtitle("Salary vs. Experience (Training Set)") + xlab('Years of Experience') + ylab('Salary')
pl3
```

```r
#Visualizing the test set results
library(ggplot2)

pl1 = ggplot() + geom_point(aes(x = test_set$YearsExperience, y = test_set$Salary), col = 'red') 
pl2 = pl1 + geom_line(aes(x = training_set$YearsExperience, y = predict(regressor, newdata = training_set)), col = 'blue')
pl3 = pl2 + ggtitle("Salary vs. Experience (Test Set)") + xlab('Years of Experience') + ylab('Salary')
pl3
```
